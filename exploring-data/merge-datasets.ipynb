```python
import pandas as pd

era_df = pd.read_csv('exploring-data/era5-germany-july-september-2022.csv')
```

```python
era_df.sample(5)
```

```python
era_df = era_df.rename(columns={'valid_time': 'Datetime (UTC)'})
```

```python
era_df.sample()
```

```python
era_df = era_df.drop(columns=['latitude', 'longitude'])
```

```python
era_df.sample()
```

```python
era_df.isnull().sum()
```

```python
price_df = pd.read_csv('exploring-data/electricity-price-july-september-2022.csv')
```

```python
price_df.sample(5)
```

```python
price_df = price_df.rename(columns={'valid_time': 'Datetime (UTC)'})
```

```python
price_df.isnull().sum()
```

```python
price_df = price_df.drop(columns=['Datetime (Local)', 'ISO3 Code'])
price_df.sample()
```

```python
pvgis_df = pd.read_csv('exploring-data/pvgis-germany-2022.csv')
```

```python
pvgis_df.sample(5)
```

```python
pvgis_df = pvgis_df.rename(columns={'time': 'Datetime (UTC)'})
```

```python
pvgis_df = pvgis_df.drop(columns=['Int'])
```

```python
pvgis_df.sample()
```

```python
pvgis_df['Datetime (UTC)'] = pd.to_datetime(
    pvgis_df['Datetime (UTC)'].astype(str),
    format='%Y%m%d:%H%M'
)
```

```python
pvgis_df
```

```python
era_df
```

```python
era_df['Datetime (UTC)'] = pd.to_datetime(era_df['Datetime (UTC)'])
price_df['Datetime (UTC)'] = pd.to_datetime(price_df['Datetime (UTC)'])
pvgis_df['Datetime (UTC)'] = pd.to_datetime(pvgis_df['Datetime (UTC)'])

pvgis_df['Datetime (UTC)'] = pvgis_df['Datetime (UTC)'].dt.floor('h')

print('era_df sample:', era_df['Datetime (UTC)'].iloc[0])
print('price_df sample:', price_df['Datetime (UTC)'].iloc[0])
print('pvgis_df sample:', pvgis_df['Datetime (UTC)'].iloc[0])
```

```python
merged = pd.merge(era_df, price_df, on='Datetime (UTC)', how='inner')
print('era + price shape:', merged.shape)

final_df = pd.merge(merged, pvgis_df, on='Datetime (UTC)', how='inner')
print('merged + pvgis shape:', merged.shape)
final_df.head()
```

```python
final_df.to_csv('exploring-data/final-dataset.csv', index=False)
```
