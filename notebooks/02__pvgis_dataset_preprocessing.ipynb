```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/pvgis-raw.csv')
```

```python
df.head()
```

```python
df['time'] = df['time'].str.replace(
    r'(\d{4})(\d{2})(\d{2}):(\d{2})(\d{2})',
    r'\1-\2-\3 \4:\5',
    regex=True
)
```

```python
df['time'].sample()
```

```python
df = df.rename(columns={'time': 'Datetime (UTC)'})
```

```python
df.head()
```

```python
df = df.drop(columns=['Int'])
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['P'])
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Gb(i)', 'Gd(i)', 'Gr(i)', 'H_sun', 'T2m', 'WS10m'])
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['P'].value_counts()
```

```python
df['Datetime (UTC)'].head(5)
```

```python
df['Datetime (UTC)'] = pd.to_datetime(df['Datetime (UTC)'])
df['Datetime (UTC)'] = df['Datetime (UTC)'] - pd.Timedelta(minutes=30)
```

```python
df['Datetime (UTC)'].head(5)
```

```python
df.shape
```

```python
df = df[(df['Datetime (UTC)'] >= '2023-07-01') &
                     (df['Datetime (UTC)'] < '2023-10-01')]
```

```python
df.shape
```

```python
df.to_csv('../data/processed/pvgis.csv', index=False)
```
