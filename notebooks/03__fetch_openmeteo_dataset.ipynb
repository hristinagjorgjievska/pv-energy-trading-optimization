```python
import requests
import pandas as pd
```

```python
url = "https://historical-forecast-api.open-meteo.com/v1/forecast"

params = {
    "latitude": 51.0,
    "longitude": 10.0,
    "start_date": "2023-07-01",
    "end_date": "2023-09-30",
    "hourly": "temperature_2m,wind_speed_10m,wind_direction_10m,cloud_cover,shortwave_radiation,precipitation",
    "timezone": "UTC"
}

response = requests.get(url, params=params)
data = response.json()
```

```python
df = pd.DataFrame(data["hourly"])
df.to_csv("meteo-raw.csv", index=False)
```
