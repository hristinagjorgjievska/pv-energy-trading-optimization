```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('../data/processed/rq1_features.csv')
```

```python
train_end = int(len(df) * 0.70)
val_end = int(len(df) * 0.85)

train = df.iloc[:train_end]
val = df.iloc[train_end:val_end]
test = df.iloc[val_end:]

print(f"Train: од {train.index.min()} до {train.index.max()} ({len(train)} редови)")
print(f"Val:   од {val.index.min()} до {val.index.max()} ({len(val)} редови)")
print(f"Test:  од {test.index.min()} до {test.index.max()} ({len(test)} редови)")
```

```python
y_true = test['Price (EUR/MWhe)']
y_pred = test['price_lag_24h']

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
r2 = r2_score(y_true, y_pred) * 100

print(f"MAE:  {mae:.2f} EUR/MWh")
print(f"RMSE: {rmse:.2f} EUR/MWh")
print(f"R²:   {r2:.3f}%")
```
