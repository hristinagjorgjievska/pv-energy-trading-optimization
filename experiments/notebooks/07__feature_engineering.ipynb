```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/final_dataset.csv')
```

```python
df.columns
```

```python
df['price_lag_1h'] = df['Price (EUR/MWhe)'].shift(1)
df['price_lag_24h'] = df['Price (EUR/MWhe)'].shift(24)
df['price_lag_168h'] = df['Price (EUR/MWhe)'].shift(168)
```

```python
df['P_lag_1h'] = df['P'].shift(1)
df['P_lag_24h'] = df['P'].shift(24)
```

```python
df = df.dropna()
```

```python
train_end = int(len(df) * 0.70)
val_end = int(len(df) * 0.85)

train = df.iloc[:train_end]
val = df.iloc[train_end:val_end]
test = df.iloc[val_end:]

print(f"Train: од {train.index.min()} до {train.index.max()} ({len(train)} редови)")
print(f"Val:   од {val.index.min()} до {val.index.max()} ({len(val)} редови)")
print(f"Test:  од {test.index.min()} до {test.index.max()} ({len(test)} редови)")
```

```python
df.to_csv('../data/processed/rq1_features.csv')
```
