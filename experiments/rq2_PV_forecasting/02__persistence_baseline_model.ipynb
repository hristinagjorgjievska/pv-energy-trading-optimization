```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
```

```python
df = pd.read_csv('../data/processed/rq2_features_daylight.csv', index_col='Datetime (UTC)', parse_dates=True)
```

```python
y_true = df['P']
y_predicted = df['P_lag_24h']
```

```python
print(f"NaN во P_lag_24h: {y_predicted.isna().sum()}")
```

```python
train_end = int(len(df) * 0.70)
validation_end = int(len(df) * 0.85)

train = df.iloc[:train_end]
validation = df.iloc[train_end:validation_end]
test = df.iloc[validation_end:]

print(f"Train: од {train.index.min()} до {train.index.max()} ({len(train)} редови)")
print(f"Test:  од {test.index.min()} до {test.index.max()} ({len(test)} редови)")
```

```python
y_true_test = test['P']
y_pred_test = test['P_lag_24h']

mae = mean_absolute_error(y_true_test, y_pred_test)
rmse = root_mean_squared_error(y_true_test, y_pred_test)
r2 = r2_score(y_true_test, y_pred_test)

print(f"MAE:  {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²:   {r2*100:.1f}%")
```
